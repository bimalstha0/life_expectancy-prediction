## Outlier handling
### The outlier seems to have made the data noisy as seen from visualizations. So the outliers are handled by imputation with the help of interquartile range.

```python
# defining function for dealing with outliers
def impute_outliers(data,col):
    df = pd.DataFrame(data)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

# Calculate the IQR (Interquartile Range)
    IQR = Q3 - Q1

# Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    fixed=[]
# Identify outliers
# Replace outliers with a suitable value (e.g., the nearest non-outlier value)
    for index, value in df[col].items():    
        if(value<lower_bound):
            value = lower_bound
        elif(value>upper_bound):
            value = upper_bound
        else:
            value=value
        fixed.append(value)
    # Create a figure and two subplots side by side
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

    # Create boxplots for the two datasets on the respective subplots
    ax1.boxplot(df[col], vert=False)
    ax2.boxplot(fixed, vert=False)

    # Set titles for the subplots
    ax1.set_title(f'{col} Before ')
    ax2.set_title(f'{col} After')

    # Display the plots
    plt.show()
    df[col]=fixed
    
for col in cols_for_norm:
    impute_outliers(cols_for_norm,col)
```

## Handling Categorical Data
### We had two categorical features in the dataset: Status and Country. For machine learning model to understand these features they need to be encoded.

### Ordinal Encoding for Status
### Ordinal encoding is a technique used in data preprocessing to convert categorical data into numerical values. This method is particularly useful when dealing with categorical variables that have an inherent order or rank. Ordinal encoding assigns a unique integer to each category, preserving the ordinal relationship between the categories.

```python
category_order = ['Developing', 'Developed']

# Create an OrdinalEncoder instance with specified category order
encoder = OrdinalEncoder(categories=[category_order])

# Fit and transform the data
encoded_data = encoder.fit_transform(df[['Status']])

# Create a new DataFrame with the encoded values
df['Status_Encoded'] = encoded_data

unique_values, counts = np.unique(df['Status_Encoded'], return_counts=True)
# Display the unique values and their counts
for value, count in zip(unique_values, counts):
    print(f"{value}: {count}")
```

### One Hot Encoding for Nominal Data
### One-hot encoding is a data preprocessing technique used to convert categorical data into a binary or numerical format that is suitable for machine learning algorithms. This method is particularly useful for categorical variables with no inherent order, where each category is considered independent like Country.

```python
# Perform one-hot encoding using pandas' get_dummies function
countries_encoded = pd.get_dummies(df.Country, columns=['Country'], prefix='Country')

# Display the one-hot encoded DataFrame
countries_encoded
```

### Combine all the columns for model training


```python
# combine all the segragated data into one for model training
final_df = pd.concat([transformed_df,df['Status_Encoded'],countries_encoded,cols_for_norm['Life expectancy']],axis=1)
final_df
```

## Utilization of Analysis Result
### Based on the analysis results, a machine learning model is built. By identifying the features and they can be used as a base for predicting life expectancy. The regression model used here was ExtraTrees.

### Extra Trees, short for Extremely Randomized Trees, is an extension of the Random Forest Algorithm. Like Random Forest, it builds a forest of decision trees but with additional randomization in tree building process. This Model was used since it is highly robust to overfitting, has improved generalization, and can identify feature importance. 


```python
# preparing data for model 
y=final_df['Life expectancy']
X=final_df.drop(columns='Life expectancy')
```

```python
# create instances for cross validation
k = 5  # Number of folds
kf = KFold(n_splits=k, shuffle=True, random_state=42)
scoring = make_scorer(mean_absolute_error) 
```

```python
# create instance for the model
model= ExtraTreesRegressor()

# Perform cross-validation
cross_val_scores = cross_val_score(model, X, y, cv=kf, scoring=scoring)
mean_score = cross_val_scores.mean()
std_deviation = cross_val_scores.std()

print(f"Mean MAE: {mean_score}")
print(f"Standard Deviation: {std_deviation}")
```

```python
# splitting data randomly into test and train
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,shuffle=True,random_state=50)

#train the model on test data
model.fit(X_train,y_train)

#predict the value on test data
y_pred = model.predict(X_test)

#calculating mean absolute error, mean squared error and r squared for the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
```
